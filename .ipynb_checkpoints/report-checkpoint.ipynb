# Social Influences on World Happiness

This data is from 2019, and while not the most recent data, I thought it would be interesting to see how to the data analyzed now might compare to data in the future post the pandemic and other major historic moments.

## Background and Problem

The dataset I used is derived from the Gallup's World Poll showcases data measured from an average of a survey asking participants to rate their happiness on a scale. Using this scale we are able to compare and contrast how on average a citizen is able to base their happiness, whether it be how the economy is thriving, how corrupt their country is, or their personal liberties for example. Using this data set, we can only determine how each of the different factors compare overall based on the country's given ranking, but does not prove any real relationship between the differing factors. 

#### Is a country's happiness based on social factors?


## Data Dictionary

| Field | Description |
| :--- | :--- |
| Overall rank | The overall happiness rank of a country based on the data from Gallup's World Poll |
| Country or region | Name of the area |
| Score | A happiness score based on the average of the results of the survey|
| GDP per capita | The GDP of the area |
| Social support | How much the community contributes to overall happiness |
| Healthy life expectancy | How the life expectancy contributed to the overall rank |
| Freedom to make life choices | How their personal liberities contributed to the overall rank |
| Generosity | How their generosity contributed to the overall rank | 
| Perceptions of corruption | How the perceptions of corruption contributed to the overall rank |


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_2019 = pd.read_csv("data/2019.csv")

data_2019.isnull().sum() # checks to see if there are any null values, and lucky for me, there isn't anything to clean
```

After importing and checking the dataset from 2019, I decided to at first observe all parts of the data. I then plotted the data with a scatterplot, focusing on the "Social Support" variable initially which allowed me to simply visualize parts of the data before diving in. I chose to plot most of the variables by social support rather than just plotting social support by happiness to be able to analyze the relationship between them, while interpreting happiness to be the resulting slope. 

```python
sns.lmplot(x = 'Freedom to make life choices', y = 'Social support', data = data_2019)
g = sns.scatterplot(x = 'Freedom to make life choices', y = 'Social support', hue = 'GDP per capita', data = data_2019)
g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

Looking at this graph initially, we can see sort of a "loose" positive correlation with freedom and social support, emphasizing the fact that the amount of freedoms and how the community supports in each other can be considered directly related. If we look at the plot on a smaller scale, we can see that the more that the higher the freedom and social support is, the GDP per capita seems to follow. I chose to add the GDP per capita in my plots to further analyze the relationship (if any) between social support and monetary happiness. 

Following the graph, I decided to look at the numbers to look at the data quanitiatively rather than visually. Reading the first few and last few rows of data based on social support, there seems to be a slight correlation with social support and freedom, but is not a precise correlation, which was reflected on the plot, and therefore the happiness score and rank were lower. 

```python
data_2019.sort_values(by='Social support', ascending=False).head()
```

```python
data_2019.sort_values(by='Social support', ascending=False).tail()
```

One thing that I noticed while looking at the first and last few rows of data is that when the social support was high, the healthy life expectancy was high and when the social support was lower, the health life expectancy was lower.

I decided to plot it following a similar method, also comparing GDP per capita to the social support and healthy life expectancy and ended up with very similar results with a stronger correlation. The higher the healthy life expectancy and social support, the more the GDP per capita can relate, which leads me to reconsider the age old question:

"Can money (or at least your country's GDP) lead to happiness?"
The answer to that is kind of, but not really. After sorting the values based on the GDP per capita and comparing the overall rank, I noticed that regardless of the GDP per capita values, the happiness score and ranking was not strictly consistent in what we normally would assume. I plotted the GDP per capita against the happiness rank just to visualize the correlation. While there seems to be a stronger correlation, there are some obvious outliers in our data.

```python
sns.lmplot(x = 'Healthy life expectancy', y = 'Social support', data = data_2019) 
g = sns.scatterplot(x = 'Healthy life expectancy', y = 'Social support', hue = 'GDP per capita', data = data_2019) 
g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
data_2019.sort_values(by='GDP per capita', ascending=False).head()
```

```python
data_2019.sort_values(by='GDP per capita', ascending=False).tail()
```

```python
sns.lmplot(x = 'Score', y = 'GDP per capita', data = data_2019)
g = sns.scatterplot(x = 'Score', y = 'GDP per capita', hue = 'Overall rank', data = data_2019)
g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

I plotted the perceptions of corruption by the social support and noticed that perceptions of corruption and social support have a slightly positive relationship (with a lot of outliers) which would make sense since the higher the corruption, the less likely you are to say you are happy. This also shows us that there isn't really a relationship between perceptions of corruption and social support.

```python
sns.lmplot(x = 'Perceptions of corruption', y = 'Social support', data = data_2019) 
g = sns.scatterplot(x = 'Perceptions of corruption', y = 'Social support', hue = 'GDP per capita', data = data_2019)
g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```
